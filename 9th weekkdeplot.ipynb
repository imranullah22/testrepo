```python
dd = datasets.load_iris()
dd_df = pd.DataFrame(dd.data, columns=['Sepal_Length','Sepal_Width', 'Patal_Length', 'Petal_Width'])    
dd_df['Species'] = dd.target 
dd_df['Species'].replace([0], 'Iris_Setosa', inplace=True)
dd_df['Species'].replace([1], 'Iris_Vercicolor', inplace=True)
dd_df['Species'].replace([2], 'Iris_Virginica', inplace=True)
dd_df
```

```python
# Plotting the KDE Plot
sns.kdeplot(dd_df.loc[(dd_df['Species']=='Iris_Virginica'),
            'Sepal_Length'], color='b', shade=True, Label='Iris_Virginica')
  
# Setting the X and Y Label
plt.xlabel('Sepal Length')
plt.ylabel('Probability Density')
```
